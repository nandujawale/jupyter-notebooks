```python
import pandas as pd
```

```python
df = pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
df.nunique()
```

```python
df.shape
```

```python
df.columns
```

```python
useful_columns = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount',
                  'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']

df = df[useful_columns]
df.sample(n=5)
```

```python
df.shape
```

```python
df.columns[df.isna().any()]
```

```python
df.isna().sum()
```

```python
fill_with_zero_columns = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[fill_with_zero_columns] = df[fill_with_zero_columns].fillna(0)
df.isna().sum()
```

```python
df['Landsize'] = df['Landsize'].fillna(df['Landsize'].mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df['BuildingArea'].mean())
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.sample(n=5)
```

```python
df = pd.get_dummies(df, drop_first=True)
df.sample(n=5)
```

```python
df.shape
```

```python
X = df.drop(columns='Price')
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_train, y_train), lr.score(X_test, y_test)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=50, tol=0.1)
lasso.fit(X_train, y_train)
lasso.score(X_train, y_train), lasso.score(X_test, y_test)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=50, tol=0.1)
ridge.fit(X_train, y_train)
ridge.score(X_train, y_train), ridge.score(X_test, y_test)
```
